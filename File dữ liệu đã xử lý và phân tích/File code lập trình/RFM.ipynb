### File code này để tính ra điểm RFM của từng khách hàng và lưu vào file `rfm.csv`

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dft = pd.read_excel("DATA SET - VÒNG 1 CUỘC THI DATA GOT TALENT 2023.xlsx", sheet_name="ticket")
print(dft.info())
dft.head(50)
```

```python
# Gộp các orderid trùng nhau
dft_clean = dft.groupby('orderid', as_index=False).agg(lambda x: x.unique()[0])
```

```python
dft_clean.head(50)
```

```python
selected_columns = ['customerid', 'saledate', 'total']
rfm_df = dft_clean.loc[:, selected_columns]
```

```python
rfm_df
```

```python
rfm_df['customerid'].value_counts()
```

### Select data 

```python
df = rfm_df.copy()
```

```python

# Lấy ngày giao dịch gần nhất của tất cả khách hàng
latest_transaction_date_all_customers = df['saledate'].max()

# Tính Recency
df['recency'] = (latest_transaction_date_all_customers - df.groupby('customerid')['saledate'].transform('max')).dt.days

# Tính Frequency
df['frequency'] = df.groupby('customerid')['customerid'].transform('count')

# Tính Monetary
df['monetary'] = df.groupby('customerid')['total'].transform('sum')

# Hiển thị kết quả RFM
rfm_result = df[['customerid', 'recency', 'frequency', 'monetary']].drop_duplicates().reset_index(drop=True)
print(rfm_result)
```

### Tính RFM

```python
rfm = rfm_result.copy()
```

```python
# Tính R score
rfm['r_score'] = pd.cut(rfm['recency'], bins=3, labels=[3, 2, 1], include_lowest=True)

# Tính F score
rfm['f_score'] = pd.cut(rfm['frequency'], bins=3, labels=[1, 2, 3], include_lowest=True)

# Tính M score
rfm['m_score'] = pd.cut(rfm['monetary'], bins=3, labels=[1, 2, 3], include_lowest=True)

# Tính RFM cell
rfm['rfm_cell'] = rfm['r_score'].astype(str) + rfm['f_score'].astype(str) + rfm['m_score'].astype(str)

# Tính RFM score
rfm['rfm_score'] = rfm['r_score'].astype(int) + rfm['f_score'].astype(int) + rfm['m_score'].astype(int)

# Hiển thị kết quả
print(rfm[['customerid', 'r_score', 'f_score', 'm_score', 'rfm_cell', 'rfm_score']])
```
