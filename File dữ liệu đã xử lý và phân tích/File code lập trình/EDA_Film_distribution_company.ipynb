```python
from datetime import datetime, timedelta

def get_dob(number_of_days):
    # Ngày tham chiếu (ngày 1 tháng 1 năm 1900)
    reference_date = datetime(1900, 1, 1)

    # Chuyển đổi
    result_date = reference_date + timedelta(days=number_of_days - 2)
#     # In ngày kết quả
#     print(result_date.strftime("%Y-%m-%d"))
    return result_date.year
dfc['YOB'] = dfc['DOB'].apply(get_dob)
```

```python
lst_country = []
for con in dff['country']:
    if pd.isna(con):
        continue
    con = con.split(", ")
    lst_country += con
    
lst_country
# Chuyển đổi danh sách thành Series
countries_series = pd.Series(lst_country)

# Tính tần suất của từng quốc gia
frequency = countries_series.value_counts()

# In tần suất
print(frequency)

```

```python
# Vẽ biểu đồ
plt.bar(frequency.index, frequency, color='purple')

# Đặt chú thích cho trục x và làm in nghiêng
plt.xticks(rotation=45, ha='right')

# Đặt tên cho trục x và trục y
plt.xlabel('country')
plt.ylabel('Frequency')

# Đặt tiêu đề cho biểu đồ
plt.title('Frequency of country')

# Hiển thị biểu đồ
plt.show()
```

```python
duration_int = dff['duration'].apply(lambda x: int(x.split(" ")[0]))
dff['duration_int'] = duration_int
```

```python
lst_duration = dff['duration'].value_counts().index.to_list()
print(lst_duration[:3])
for i in range(len(lst_duration)):
    lst_duration[i] = int(lst_duration[i][:-4])
print(lst_duration[:3])
```

```python
count_duration = dff['duration'].value_counts()
mean_duration = 0
for i in range(len(lst_duration)):
    mean_duration += lst_duration[i] * count_duration[i]
mean_duration = mean_duration / len(lst_duration)
mean_duration
```
